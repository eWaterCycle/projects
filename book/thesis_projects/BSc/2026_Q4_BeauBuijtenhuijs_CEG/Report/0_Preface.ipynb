# Preface

Presented before you is the bachelor thesis “Impact of climate change on the Okavango River.” It has been written to complete the bachelor Civil Engineering at Delft University of Technology. I wrote this thesis because of my interest in nature and my concern about the impact of climate change on it. Because of climate change, the hydrology of the Okavango River changes, which affects the Okavango Basin and the functioning of its ecosystem. This case stood out for me due to my interests in the continent Africa. 
During my time working on this bachelor thesis, I learned a lot about hydrological modelling. My coding skills improved a lot while working on the HBV model and results analysis.
