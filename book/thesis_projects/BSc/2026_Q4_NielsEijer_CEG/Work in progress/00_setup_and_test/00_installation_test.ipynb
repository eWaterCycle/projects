```python
import xarray as xr
from pathlib import Path

staticmaps_file = Path(parameter_set.directory) / "staticmaps.nc"

ds = xr.open_dataset(staticmaps_file)

print(ds)
```

```python
lat_lobith = 51.86
lon_lobith = 6.12

nearest = ds.sel(lat=lat_lobith, lon=lon_lobith, method="nearest")

print(nearest.lat.values)
print(nearest.lon.values)
```

```python
for var in ds.data_vars:
    print(var)
```

```python
lat_model = nearest.lat.values
lon_model = nearest.lon.values

print("lat:", lat_model)
print("lon:", lon_model)

print("catchment:", ds["wflow_catchment.map"].sel(lat=lat_model, lon=lon_model).values)
print("river:", ds["wflow_river.map"].sel(lat=lat_model, lon=lon_model).values)
print("streamorder:", ds["wflow_streamorder.map"].sel(lat=lat_model, lon=lon_model).values)
print("subcatch:", ds["wflow_subcatch.map"].sel(lat=lat_model, lon=lon_model).values)
```
