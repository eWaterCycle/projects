# Voorwoord

Voor u ligt mijn bachelor thesis met de titel ”Bevaarbaarheid van de IJssel: een kwantitatieve analyse
van de invloed van klimaatverandering onder low-flow conditions”. Dit onderzoek is geschreven ter
afronding van mijn bachelor Civiele Techniek aan de Technische Universiteit Delft.
Tijdens mijn studie heb ik een groeiende interesse ontwikkeld in hydrologie, riviersystemen en de manier waarop klimaatverandering invloed heeft op waterbeheer. De IJssel vormt daarbij een interessant
en relevant studiegebied, niet op de laatste plaats omdat ik in de regio geboren ben. De rivier speelt
een belangrijke rol voor scheepvaart en zoetwatervoorziening, maar is tegelijkertijd gevoelig voor lage
afvoeren. In dit onderzoek heb ik daarom gekeken naar hoe toekomstige lage afvoercondities bij Lobith
kunnen doorwerken naar de bevaarbaarheid van de IJssel.
Ik wil graag mijn begeleiders dr. ir. Rolf Hut en dr. ir. Astrid Blom bedanken voor hun begeleiding,
kritische feedback en inhoudelijke ondersteuning tijdens dit project. Ook wil ik ir. Mark Melotto bedanken voor zijn ondersteuning op softwaregebied. Hun kennis en advies hebben me geholpen om het
onderzoek scherper en systematischer uit te werken. Daarnaast wil ik iedereen bedanken die heeft
meegedacht of feedback heeft gegeven tijdens het proces.
Ik hoop dat dit onderzoek bijdraagt aan een beter begrip van de mogelijke gevolgen van klimaatverandering voor lage-afvoercondities en de bevaarbaarheid van de IJssel.

Niels Eijer
Delft, Juni, 2026
