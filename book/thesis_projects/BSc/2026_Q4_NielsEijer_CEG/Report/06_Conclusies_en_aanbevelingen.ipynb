# Hoofdstuk 6 - Conclusies en aanbevelingen

## 6.1 Beantwoording van de deelvragen

### Hoe veranderen lage afvoercondities bij Lobith onder toekomstige klimaatverandering?

De resultaten laten zien dat lage afvoercondities bij Lobith onder alle drie de klimaatscenario's verslechteren ten opzichte van de historische referentieperiode (1990-2019), maar in zeer verschillende mate. Onder SSP1-2.6 blijft de verandering beperkt: het q5-percentiel daalt licht van 1.057 m3/s naar 1.038 m3/s en het gemiddelde aantal dagen onder de OLA-drempel van 1.020 m3/s stijgt van 14,3 naar 15,8 dagen per jaar. Onder SSP2-4.5 en vooral SSP5-8.5 is de verschuiving veel groter: het q5-percentiel daalt tot respectievelijk 969 m3/s en 945 m3/s, en het aantal kritieke laagwaterdagen stijgt van 27,4 naar 33,5 dagen per jaar. De respons is dus duidelijk niet-lineair: de stap van een laag- naar een midden-impactscenario levert in relatieve termen een grotere verslechtering op dan de stap van een midden- naar een hoog-impactscenario, wat erop wijst dat met name het voorkomen van forse klimaatmitigatie, zoals in SSP1-2.6, een groot deel van de laagwaterproblematiek kan voorkomen.

### Hoe kunnen deze lage afvoercondities worden gekoppeld aan de bevaarbaarheid op de IJssel?

In dit onderzoek is de koppeling tussen de lage-afvoercondities en de bevaarbaarheid gelegd via de Overeengekomen Lage Afvoer (OLA) van 1.020 m3/s bij Lobith, die wordt gebruikt als hydrologische indicator voor verhoogde laagwaterdruk op het systeem. Omdat de uiteindelijke waterbeschikbaarheid op de IJssel niet alleen van de Lobithafvoer afhangt, maar ook van de afvoerverdeling bij de Pannerdense Kop en de IJsselkop, is deze koppeling aangevuld met twee gevoeligheidsvarianten, een 10%- en een 17%-scenario, die een mogelijke verschuiving van de afvoerverdeling ten gunste van de Waal simuleren. Deze aanpak maakt het mogelijk om niet alleen de klimaatgedreven verandering in lage afvoer te kwantificeren, maar ook om in te schatten hoe gevoelig de uitkomsten zijn voor onzekerheid in de afvoerverdeling. De koppeling is daarmee een eerste-orde-indicator van bevaarbaarheidsrisico, geen directe voorspelling van lokale vaardiepte.

### Hoe verandert de frequentie van perioden waarin de bevaarbaarheid van de IJssel onder druk staat onder toekomstige klimaatscenario's?

De frequentie van kritieke laagwaterdagen neemt in alle scenario's en bij alle drie de drempelwaarden toe. Bij de OLA-drempel stijgt het jaarlijkse aantal kritieke dagen van 14,3 historisch naar 15,8 onder SSP1-2.6; 27,4 onder SSP2-4.5 en 33,5 onder SSP5-8.5. Wanneer de gevoeligheid voor de afvoerverdeling wordt meegenomen, wordt dit effect versterkt: bij het 17%-scenario loopt het jaarlijkse aantal kritieke dagen onder SSP5-8.5 op tot 70,1, bijna twee keer zoveel als het historische gemiddelde bij de OLA-drempel. Ook de spreiding tussen de jaren verandert: waar in de historische periode 19 van de 30 jaren geen enkele kritieke laagwaterdag telden, is dit onder SSP5-8.5 in geen van de 30 gemodelleerde jaren het geval. Laagwater verschuift daarmee van een incidenteel verschijnsel naar een structureel terugkerend kenmerk van het systeem.

### Hoe verandert de duur van aaneengesloten perioden waarin de bevaarbaarheid van de IJssel onder druk staat onder toekomstige klimaatscenario's?

Naast de frequentie neemt ook de duur van aaneengesloten kritieke laagwaterperioden toe. Het gemiddelde aantal aaneengesloten periodes stijgt van 35 in de historische referentie naar 57 onder SSP5-8.5, met een toename van gemiddelde duur van 12,3 naar 17,6 dagen. Het meest uitgesproken is de verandering in de maximale duur: waar de langste aaneengesloten laagwaterperiode in de historische reeks 38 dagen bedraagt, loopt dit onder SSP5-8.5 op tot 120 dagen. Deze toename in duur is voor de scheepvaart relevanter dan een gelijk aantal verspreide losse dagen, omdat langdurige beperkingen de transportplanning en betrouwbaarheid van de vaarweg sterker aantasten.

## 6.2 Beantwoording van de hoofdvraag

De hoofdvraag van dit onderzoek luidde in welke mate klimaatverandering de frequentie en duur van perioden met onvoldoende vaardiepte op de IJssel beïnvloedt, op basis van klimaatgedreven veranderingen in de lage afvoer bij Lobith. Op basis van de uitgevoerde analyse kan worden geconcludeerd dat klimaatverandering een aanzienlijke en scenarioafhankelijke invloed heeft op deze laagwaterdruk. Onder een laag-impactscenario (SSP1-2.6) blijft de toename beperkt tot een stijging van ongeveer 10% in het aantal kritieke laagwaterdagen ten opzichte van de historische situatie. Onder een midden- en hoog-impactscenario (SSP2-4.5 respectievelijk SSP5-8.5) is de toename veel substantiëler, met stijgingen van 91% en 134% in het aantal kritieke dagen bij de OLA-drempel. Wanneer bovendien wordt aangenomen dat de afvoerverdeling zich ten gunste van de Waal blijft ontwikkelen, zoals eerdere studies suggereren, kan dit effect verder oplopen tot bijna een verdubbeling of verdrievoudiging van het historische gemiddelde. De duur van aaneengesloten laagwaterperioden volgt een vergelijkbaar patroon en bereikt onder het hoog-impactscenario waarden die het historische maximum ruim overschrijden. Klimaatverandering vergroot daarmee zowel de frequentie als de duur van perioden waarin de bevaarbaarheid van de IJssel onder druk staat, met een effect dat sterk afhangt van het emissiepad en van de mate waarin de afvoerverdeling binnen de Rijndelta in de toekomst verschuift.

## 6.3 Conclusie

Dit onderzoek laat zien dat de bevaarbaarheid van de IJssel onder toekomstige klimaatverandering kwetsbaarder wordt, en dat deze kwetsbaarheid door twee, deels onafhankelijke, onzekerheden wordt bepaald: de mate van klimaatverandering zelf en de ontwikkeling van de afvoerverdeling binnen de Rijndelta. Voor de afvoer bij Lobith geldt dat midden- en hoog-impactscenario's leiden tot een duidelijke en aanhoudende toename van het aantal kritieke laagwaterdagen, van 14,3 dagen per jaar historisch naar 27,4 respectievelijk 33,5 dagen per jaar, en de duur van aaneengesloten laagwaterperioden, van gemiddeld 12,3, naar 15,8 respectievelijk 17,6 dagen, met een maximale duur die toeneemt van 38 naar 120 dagen. Daarnaast laat de gevoeligheidsanalyse zien dat de keuze voor de afvoerverdeling richting de IJssel een invloed heeft die in dezelfde orde van grootte ligt als het verschil tussen de klimaatscenario's: bij een veronderstelde 17% afname van het IJsselaandeel stijgt het aantal kritieke dagen onder SSP5-8.5 tot 70,1 per jaar, tegenover 33,5 bij ongewijzigde verdeling. Dit betekent dat onzekerheid over de afvoerverdeling minstens zo bepalend is voor de toekomstige bevaarbaarheid van de IJssel als de onzekerheid over het klimaatscenario zelf. Tegelijkertijd onderstrepen de resultaten dat het hier gaat om een eerste-orde-inschatting van systeembrede laagwaterdruk: de daadwerkelijke vaardiepte op de IJssel wordt mede bepaald door het stuwbeheer bij Driel, bodemligging, sedimentatie en lokale vaargeulafmetingen die in dit onderzoek niet afzonderlijk zijn doorgerekend. De waarde van dit onderzoek ligt dan ook vooral in het blootleggen van de richting en orde van grootte van de verandering, en in de bevestiging dat een relatief eenvoudig, reproduceerbaar modelinstrumentarium daarvoor een bruikbaar beeld kan opleveren dat aansluit bij de richting van bestaande, uitgebreidere studies.

## 6.4 Aanbevelingen voor vervolgonderzoek

Voor vervolgonderzoek is het wenselijk om de afvoerverdeling bij de Pannerdense Kop en de IJsselkop niet langer via vaste gevoeligheidsdrempels te benaderen, maar expliciet hydraulisch te modelleren, bijvoorbeeld door de koppeling van het wflow-model aan een hydraulisch model van de bifurcatiepunten. Hierdoor zou de invloed van klimaatverandering op de afvoerverdeling zelf kunnen worden gekwantificeerd in plaats van verondersteld. Daarnaast verdient de gebruikte delta-change-methode aandacht: deze simuleert geen geleidelijke overgang tussen heden en het einde van de eeuw en is afgeleid van één klimaatsignaal per SSP-scenario. Het gebruik van meerdere klimaatmodellen per SSP-scenario zou de onzekerheidsband binnen elk scenario beter in beeld brengen. Daarnaast kan tijdafhankelijke scenarioforcing, waarbij klimaatsignalen geleidelijk veranderen richting het einde van de eeuw, inzicht geven in de ontwikkeling van de laagwaterdruk in de tussenliggende decennia. Ook kan de seizoensafhankelijke biascorrectie worden verfijnd: deze corrigeert primair de afvoerverdeling, maar niet de timing van afvoergebeurtenissen, terwijl juist de timing bepalend is voor de duur van aaneengesloten laagwaterperioden. Vervolgonderzoek zou daarom kunnen onderzoeken of biascorrectiemethoden die ook autocorrelatie en seizoensvolgorde behouden een betrouwbaarder beeld geven van de duurstatistiek. Tot slot zou een directe koppeling tussen de gesimuleerde Lobithafvoer en een lokaal hydraulisch model van de IJssel, inclusief stuwbeheer bij Driel en bodemligging, de vertaalslag van hydrologische laagwaterdruk naar daadwerkelijke vaardiepte aanzienlijk kunnen verbeteren.

## 6.5 Aanbevelingen voor toepassing in waterbeheer en scheepvaart

Voor waterbeheerders en de scheepvaartsector laten de resultaten zien dat laagwater op de IJssel zich onder een midden- of hoog-emissie klimaatscenario ontwikkelt van een incidenteel naar een structureel terugkerend verschijnsel. Dit pleit voor een waterbeheer dat niet alleen reageert op individuele droge jaren, maar rekening houdt met structurele toename van zowel de frequentie als de duur van laagwaterperioden, vooral met het oog op de inzet van de stuw bij Driel, die bij Lobithafvoeren onder 1.600 m3/s nog maar beperkt kan compenseren. Voor de scheepvaartsector is vooral de toename van de maximale duur van aaneengesloten laagwaterperioden relevant: een periode van enkele maanden, zoals onder SSP5-8.5 gevonden, vraagt om een andere operationele aanpak dan een kortdurende beperking, bijvoorbeeld in de vorm van aangepaste beladingsstrategieën, alternatieve routering via de Waal of tijdige communicatie van verwachte laagwaterperioden. Daarnaast onderstrepen de resultaten het belang van het meenemen van de onzekerheid in de afvoerverdeling in toekomstig beleid: omdat deze onzekerheid minstens zo bepalend is als de keuze van het klimaatscenario is voor de robuustheid van investeringsbeslissingen in de IJssel, zoals baggeronderhoud of aanpassing van het vaarwegprofiel, van belang om niet alleen met klimaatverandering rekening te houden, maar ook met scenario's voor veranderingen in de afvoerverdeling.
