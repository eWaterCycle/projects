## 8 Conclusie en Aanbevelingen

Dit onderzoek is begonnen met de hoofdvraag;
Wat is het effect van klimaatverandering op de kritische hoogte in het Brokopondo
stuwmeer zodat de Afobaka waterkrachtcentrale genoeg stroom kan leveren?
Het is te zeggen dat de toekomst van het Brokopondomeer er niet positief uit ziet als er
niks verandert aan het gebruik van het water. Zolang dezelfde afvoer van 200 m3
/s wordt
toegepast zal het meer na 2051 onder de kritische hoogte vallen en onbruikbaar zijn
door een te lage levering van vermogen.
SSP1.2-6 en SSP2.4-5 hebben allebei meer dan 30 droge dagen in elk jaar met droge
dagen. Dit betekent dat in het geval dat er zeer kritische droogte ontstaat dit voor langere
tijd zal aanhouden en een groot probleem zal vormen. Het aantal dagen onder de
kritische en zeer kritische hoogte neemt sterk toe in de toekomst in alle drie de
klimaatscenario’s.
Wat opvallend is, is dat de SSP1.2-6 minder neerslag heeft dan de SSP2.4-5 over vrijwel
de gehele toekomstvoorspelling wat frappant is gezien het de meest positieve
toekomstvoorspelling zou moeten zijn.
Tot 2030 is er genoeg water in het meer zoals te zien in de resultaten, maar daarna is het
niet meer mogelijk volgens de voorspellingen om genoeg vermogen te leveren. Het is wel
op te merken dat de SSP1.2-6 en de SSP2.4-5 nog redelijk dicht in de buurt liggen van de
zeer kritische hoogte en tussen 2030 en 2050 relatief stabiel liggen. Het zou mogelijk een
oplossing zijn om in deze periode de afvoer van de Afobakadam lager te zetten en
minder elektriciteit te genereren op momenten dat er minder vraag is, zoals in de nacht.
Als SSP5.8-5 een realiteit wordt dan is het nauwelijks mogelijk de dam te gebruiken. Er
zullen een stuk minder buien vallen en deze buien zullen in lager volume zijn dan in het
verleden waardoor de dam niet aangevuld kan worden.
Na 2050 lijkt het lastig te zijn om de dam te laten werken met de huidige afvoer. De enige
mogelijkheden om de dam bruikbaar te houden is om het debiet te verlagen en te
accepteren dat er minder energie opgewekt kan worden. Het voordeel hieraan is dat het
meer nog hoog genoeg zal staan om bij grotere buien weer tijdelijk meer energie op te
kunnen wekken. De potentiële energie kan dan benuttigd worden op momenten dat er
grote vraag naar is, in plaats van zo veel mogelijk er uit te halen in korte tijd en niet naar
de toekomst te kijken.
