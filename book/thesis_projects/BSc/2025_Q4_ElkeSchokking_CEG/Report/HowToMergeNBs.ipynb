```python
import nbformat

def merge_notebooks(filenames, output_filename):
    merged = nbformat.v4.new_notebook()
    merged.cells = []

    for fname in filenames:
        with open(fname, 'r', encoding='utf-8') as f:
            nb = nbformat.read(f, as_version=4)
            merged.cells.extend(nb.cells)

    with open(output_filename, 'w', encoding='utf-8') as f:
        nbformat.write(merged, f)

merge_notebooks([
    "ABSTRACT-Copy1.ipynb",
    "PREFACE.ipynb",
    "SUMMARY.ipynb",
    "TableOfContents.ipynb",
    "NOMENCLATURE-Copy1.ipynb",
    "INTRODUCTION_CH1-Copy1.ipynb",
    "BACKGROUND_CH2-Copy1.ipynb",
    "HISTORIC_LowFlow_CH3-Copy1.ipynb",
    "FUTURE_LowFlow_CH4.ipynb",
    "REFERENCES-Copy1.ipynb"
], "TUSSENRAPPORT_EFS.ipynb")
```
