# Appendix A: Snow depletion curve

Figure 16 below shows the snow depletion curve for the Upper Athabasca River (UAR) region above Hinton, covering around 10,000 km2 (Siemens et al., 2021). This shows how snowmelt may lead to sudden peaks in discharge when the snow cover depletes quickly. Although the figure is based on the UAR, the vast majority of the LAR basin lies below 1500 m elevation and may behave similarly. This line will therefore have the biggest contribution to discharge into the LAR: The snow cover depletes around the end April to start of May which is when freshet starts and aligns with the freshet algorithm (see chapter 3).
