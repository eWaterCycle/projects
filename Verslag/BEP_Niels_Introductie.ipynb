## 1.1 Aanleiding van het onderzoek

Klimaatverandering verandert het afvoerregime van rivieren. Dit gebeurt onder andere door hydrologische veranderingen in neerslag, temperatuur, verdamping, sneeuwopslag en de bijdrage van gletsjers. Voor de Rijn is dit belangrijk, omdat de afvoer bij Lobith de bovenstroomse hydrologische randvoorwaarde vormt voor de Nederlandse Rijntakken. Wanneer lage afvoercondities bij Lobith veranderen, kan dit dus ook gevolgen hebben voor de waterbeschikbaarheid op de Waal, de Nederrijn-Lek en de IJssel. Recente klimaatscenario's zoals de KNMI'23-afvoerscenario's, worden daarom gebruikt om te onderzoeken hoe het Rijnsysteem reageert op veranderende meteorologische omstandigheden en wat dat kan betekenen voor waterbeheer, scheepvaart en zoetwatervoorziening \parencite{Buitink2023}.

Binnen dit systeem is de IJssel een relatief kwetsbare tak \parencite{Rijkswaterstaat2021}. De rivier ontvangt slechts een klein aandeel van de totale afvoer bij Lobith, maar vervult een belangrijke functie binnen de regionale scheepvaart. Ook is de IJssel van uiterst belang voor de zoetwatertoevoer naar het IJsselmeer. Tijdens perioden met lage afvoer dalen de waterstanden en neemt de beschikbare vaardiepte af. Schepen kunnen dan minder diep laden, of zelfs vastlopen. Dit zorgt ervoor dat de betrouwbaarheid en de capaciteit van de vaarweg onder druk kan komen te staan. Daarmee vormt de IJssel een geschikte casus om te onderzoeken hoe klimaatgedreven veranderingen in lage afvoeren kunnen doorwerken naar een concreet civieltechnisch probleem: de bevaarbaarheid van een rivier.


## 1.2 Relevantie van het onderzoek

De maatschappelijke relevantie van dit onderzoek ligt in de kwetsbaarheid van de IJssel als vaarweg onder lage afvoercondities. Voor het hoofvaarwegennet is het van belang dat scheepvaart vlot, veilig en betrouwbaar kan plaatsvinden \parencite{Rijkswaterstaat2021}. Klimaatverandering, droogtes en bodemerosie kunnen leiden tot knelpunten in het beschikbare vaarwaterprofiel. Hierdoor kan scheepvaart vaker te maken krijgen met diepgangbeperkingen \parencite{Rijkswaterstaat2021}. Voor de IJssel is deze problematiek bijzonder relevant, omdat deze rivier door Rijkswaterstaat wordt beschreven als kwetsbaar voor onvoldoende diepte bij lage rivierafvoeren \parencite{Rijkswaterstaat2021}.

Daarnaast is het onderzoek inhoudelijk relevant, omdat bestaande studies vooral inzicht geven in veranderingen in afvoer, afvoerverdeling en morfologische ontwikkeling van de Rijndelta, terwijl de vertaling naar concrete gevolgen voor de bevaarbaarheid van de IJssel nog beperkt is uitgewerkt (\cite{Buitink2023}; \cite{Chowdhury2023}; \cite{Chowdhury2026}). De IJssel zal volgens de verwachtingen in de toekomst relatief minder afvoer verwachten \parencite{Chowdhury2026}. Dit onderzoek koppelt dergelijke hydrologische veranderingen aan een praktische indicator voor de scheepvaart en de klimaatverandering bovenstrooms van Lobith.


## 1.3 Probleemstelling

Eerdere studies hebben klimaatverandering in het stroomgebied van de Rijn vertaald naar toekomstige afvoerreeksen bij Lobith \parencite{Buitink2023}. Daarnaast laten andere onderzoeken zien dat klimaatverandering de afvoerverdeling binnen de Rijndelta kan beïnvloeden, waarbij de IJssel onder lage afvoercondities relatief minder afvoer kan ontvangen \parencite{Chowdhury2026}. Tegelijkertijd is bekend dat de IJssel gevoelig is voor lage afvoeren: klimaatverandering, droogte en bodemerosie kunnen volgens Rijkswaterstaat leiden tot knelpunten in het vaarwegprofiel en vaker tot diepgangbeperkingen\parencite{Rijkswaterstaat2021}.

Wat nog beperkt is uitgewerkt, is hoe deze toekomstige lage afvoer kan worden vertaald naar een concrete maat voor de bevaarbaarheid op de IJssel. Dit onderzoek gebruikt daarom een vereenvoudigde, reproduceerbare modelaanpak om de frequentie en duur van perioden waarin de bevaarbaarheid onder druk staat kwantitatief inzichtelijk te maken. Ook zal er gekeken worden naar hoe een relatief eenvoudig model zoals dit in de buurt komt van eerdere studies. Aan de hand van eenvoudigere modellen kunnen eenvoudiger en sneller meerdere toekomstscenario's doorgerekend worden.


## 1.4 Doelstelling
Het doel van dit onderzoek is om te bepalen in welke mate toekomstige lage-afvoercondities als gevolg van klimaatverandering leiden tot een toename in de frequentie en duur van bevaarbaarheidsproblemen op de IJssel, uitgedrukt in het aantal dagen met onvoldoende waterdiepte. Hiervoor worden met een vereenvoudigde, reproduceerbare modelaanpak toekomstige afvoercondities geanalyseerd en vertaald naar indicatoren voor bevaarbaarheidsproblemen. De nadruk ligt daarbij op de frequentie en duur van perioden waarin de bevaarbaarheid van de IJssel onder druk staat.


## 1.5 Scope
Het onderzoek richt zich op de IJssel als vaarweg binnen de Rijndelta en focust op lage-afvoercondities. De nadruk ligt op bevaarbaarheid in termen van beschikbare waterdiepte. Andere mogelijke gevolgen van lage afvoer, zoals economische schade, ecologische effecten of bredere waterkwaliteitsvraagstukken, vallen buiten de directe scope van dit onderzoek. Voor de hydrologische analyse wordt gebruikgemaakt van eWaterCycle als modelmatige basis voor het simuleren of analyseren van afvoercondities onder verschillende klimaatscenario’s. Hiervoor is vooral de klimaatverandering die zich voordoet bovenstrooms van Lobith relevant.


## 1.6 Onderzoeksvraag en deelvragen
### 1.6.1 Onderzoeksvraag

In welke mate beïnvloedt klimaatverandering de frequentie en duur van perioden met onvoldoende vaardiepte op de IJssel, op basis van klimaatgedreven veranderingen in de lage afvoer, uitgedrukt in dagen beperking per jaar voor verschillende klimaatscenario's?


### 1.6.2 Deelvragen

1. Hoe veranderen lage afvoercondities bij Lobith onder toekomstige klimaatverandering?

2. Hoe kunnen deze lage afvoercondities gelinkt worden aan bevaarbaarheid op de IJssel?

3. Hoe verandert de frequentie van perioden waarin de bevaarbaarheid van de Ijssel onder druk staat onder toekomstige klimaatscenario's?

4. Hoe verandert de duur van aaneengesloten perioden waarin de bevaarbaarheid van de Ijssel onder druk staat onder toekomstige klimaatscenario's?


## 1.7 Leeswijzer

Dit rapport is als volgt opgebouwd. Na deze introductie wordt in hoofdstuk 2 de systeem en literatuurachtergrond behandeld. Hierin wordt in detail de Nederlandse Rijndelta beschreven als gereguleerd systeem, met aandacht voor de Pannerdense kop, de IJsselkop en de historische ontwikkelingen van de afvoerverdeling. Vervolgens wordt er ingegaan op de invloed van klimaatverandering op lage afvoeren en de gevolgen daarvan op de bevaarbaarheid van de IJssel.

Hoofdstuk 3 beschrijft hoe het onderzoek uitgevoerd zal worden. Hier wordt het gebruik van eWaterCycle toegelicht en wordt in grote lijnen het model wflow uitgelegd. Ook worden de gebruikte klimaatscenario's beschreven en zal de het gebruik van de afvoerverdeling beschreven worden.

In hoofdstuk 4 worden de resultaten gepresenteerd. Eerst moet er beoordeeld worden hoe goed het wflow model overeenkomt met de waarnemingen bij Lobith. Daarna wordt onderzocht hoe lage afvoercondities veranderen aan de hand van  de verschillende klimaatscenario's en wat dit betekent voor het aantal kritieke laagwaterdagen en de duur hiervan.

In hoofdstuk 5 zullen de resultaten worden bediscussieerd. Hier zal de betekenis en de betrouwbaarheid bekeken worden. De uitkomsten worden geïnterpreteerd in relatie met de bevaarbaarheid op de IJssel. Ook zal er een vergelijking worden gelegd met de bestaande literatuur. 

Tot slot worden in hoofdstuk 6 de deelvragen en de hoofdvraag beantwoord. Aan de hand van dat antwoord zal een conclusie worden geformuleerd en worden er aanbevelingen gedaan voor eventuele vervolgonderzoeken.
